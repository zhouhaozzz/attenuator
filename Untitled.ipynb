```python
#定义随时间变化的热流密度 q(t)
def gauss(t):
    return np.exp(-4 * np.log(2) * ((t - t0)**2) / FWHM**2)

def Ne_X(N_gamma, cross_data, rho, delta_z):
    N = N_gamma * cross_data / np.sum(cross_data) * (1 - np.exp(-np.sum(cross_data) * rho * delta_z))
    return N

def Ne_Ei(Pe, Ne_X):
    N = Pe * Ne_X
    return N
```

```python
auger_Ei_1s = np.array([2630.98, 2687.08, 2974.80, 2962.85, 2743.17, 3003.90, 3018.95, 3264.62, 3279.67, 3294.72])
auger_Ei_2s = np.array([16.25, 31.42, 257.55, 272.73, 287.9])
auger_Ei_2p = np.array([199.13, 214.27, 229.42]) #eV  #俄歇电离电子能量list
auger_rate_1s = np.array([1.38e-3, 4.65e-3, 3.12e-4, 4.55e-4, 1.20e-2, 4.63e-4, 2.09e-3, 1.75e-5, 4.55e-5, 9.21e-5]) / 2.418e-17  #s^-1  #俄歇电离电子速率
auger_rate_2s = np.array([5.68e-2, 7.78e-2, 7.75e-4, 3.8e-3, 1.12e-4]) / 2.418e-17  #s^-1  #俄歇电离电子速率
auger_rate_2p = np.array([6.87e-5, 1.32e-3, 4.41e-3]) / 2.418e-17  #s^-1  #俄歇电离电子速率
auger_rate_1s_total = np.sum(auger_rate_1s)
auger_rate_2s_total = np.sum(auger_rate_2s)
auger_rate_2p_total = np.sum(auger_rate_2p)
auger_rate_1s_normalize = auger_rate_1s / auger_rate_1s_total
auger_rate_2s_normalize = auger_rate_2s / auger_rate_2s_total
auger_rate_2p_normalize = auger_rate_2p / auger_rate_2p_total


# 高斯脉冲
I0 = 3e18            # 峰值光通量
FWHM = 5e-12          # 脉冲半高宽 (s)
t0 = 5e-12              # 脉冲中心时间 (s)

total_time = 10e-12          # 总模拟时间 (s)
dt = 1e-16
time_steps = int(total_time / dt)
print_step = int(time_steps / 10)
print(time_steps)


rho = 5.28e18
delta_z = 1e-6
```

```python
NeX = np.zeros(5) #光电离电子个数
NArX = np.zeros(5) #拥有空位X的原子个数
NeEi = np.zeros(18) #俄歇电离电子数
t = 0

I = I0 / time_steps
N_atom = np.zeros(1323)
N_atom[-1] = rho

for n in range(time_steps+1):
    if n % print_step == 0:
        print(['t = ' + str(int(t / 1e-15)) + ' (fs) ' + ' total time = ' + str(int(time_steps*dt / 1e-15)+1)])
    
    I = I0 * gauss(t)
    NeX = Ne_X(I, cross_10keV, rho, delta_z) #光电离过程 假设一个时间步同一个原子只发生一次光电离

    NArX += NeX #X空位的原子个数等于光电离电子数
    
    #俄歇衰变
    NeEi_1s_total = NArX[0] * (1 - np.exp(-auger_rate_1s_total * dt)) #X=0空位 俄歇衰变次数
    NeEi_1s = NeEi_1s_total * auger_rate_1s_normalize #每个反应道衰变次数
    NArX[0] -= NeEi_1s_total #X=0空位 粒子数
    NArX[1] += NeEi_1s[0] + (NeEi_1s[1] + NeEi_1s[2] + NeEi_1s[3]) / 2
    NArX[2] += NeEi_1s[4] + (NeEi_1s[5] + NeEi_1s[6]) / 2
    NArX[3] += NeEi_1s[7] + (NeEi_1s[2] + NeEi_1s[5] + NeEi_1s[8]) / 2
    NArX[4] += NeEi_1s[-1] + (NeEi_1s[3] + NeEi_1s[6] + NeEi_1s[8]) / 2
    
    NeEi_2s_total = NArX[1] * (1 - np.exp(-auger_rate_2s_total * dt))
    NeEi_2s = NeEi_2s_total * auger_rate_2s_normalize
    NArX[1] -= NeEi_2s_total
    NArX[2] += (NeEi_2s[0] + NeEi_2s[1]) / 2
    NArX[3] += (NeEi_2s[0] + NeEi_2s[2] + NeEi_2s[3]) / 2
    NArX[4] += NeEi_2s[4] + (NeEi_2s[1] + NeEi_2s[3]) / 2
    
    
    NeEi_2p_total = NArX[2] * (1 - np.exp(-auger_rate_2p_total * dt))
    NeEi_2p = NeEi_2p_total * auger_rate_2p_normalize
    NArX[2] -= NeEi_2p_total
    NArX[3] += NeEi_2p[0] + (NeEi_2p[1]) / 2
    NArX[4] += NeEi_2p[2] + (NeEi_2p[1]) / 2
    
    
    NeEi[0:10] += NeEi_1s #记录各能量电子个数
    NeEi[10:15] += NeEi_2s
    NeEi[15:18] += NeEi_2p
    
    t = t + dt
```

```python
confi_string = ["1s", "2s", "2p", "3s", "3p", "4s", "3d", "4p", "5s", "4d", "5p", "6s", "4f", "5d", "6p", "7s", "5f", "6d", "7p"]
def read_pcs_data(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        
    data = lines[0].split()
    N1 = int(data[0])
    N2 = int(data[1])
    cross_data = np.zeros(N2)
    bind_data = np.zeros(N2)
    ke_data = np.zeros(N2)

    i = 0    
    for line in lines[1 : N1 + 1]:
        data = line.split()
        cross_data[i] = float(data[2]) * 1e-18
        i += 1
        
    i = 0    
    for line in lines[N1 + 2 : N1 + N2 + 2]:
        data = line.split()
        bind_data[i] = -float(data[3])
        ke_data[i] = float(data[4])
        i += 1

    return cross_data, bind_data, ke_data

def read_atm_data(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        
    data = lines[0].split()
    N1 = int(data[0])
    N2 = int(data[1])
    au_config = np.zeros([N1,3], dtype=int)
    au_rate = np.zeros(N1)
    au_ke = np.zeros(N1)

    i = 0    
    for line in lines[1 : N1 + 1]:
        data = line.split()
        au_config[i][0] = int(confi_string.index(data[0]))
        au_config[i][1] = int(confi_string.index(data[2]))
        au_config[i][2] = int(confi_string.index(data[3]))
        au_ke[i] = float(data[5])
        au_rate[i] = float(data[6]) / 2.418e-17
        if (data[7] == "*"):
            au_rate[i] *= int(data[8])
        i += 1
    
    fl_config = np.zeros([N2,2], dtype=int)
    fl_rate = np.zeros(N2)
    fl_ke = np.zeros(N2)
    i = 0    
    for line in lines[N1 + 2 : N1 + N2 + 2]:
        data = line.split()
        fl_config[i][0] = int(confi_string.index(data[0]))
        fl_config[i][1] = int(confi_string.index(data[2]))
        fl_ke[i] = float(data[4])
        fl_rate[i] = float(data[5]) / 2.418e-17
        if (data[6] == "*"):
            fl_rate[i] *= int(data[7])
        i += 1

    return au_config, au_rate, au_ke, fl_config, fl_ke, fl_rate

def list_files_in_directory(directory):
    return [f for f in os.listdir(directory) if os.path.isdir(os.path.join(directory, f))]

def natural_sort_key(file_name):
    # 提取文件名中的数字部分
    return [int(num) if num.isdigit() else num for num in re.findall(r'\d+|\D+', file_name)]

def count_specific_elements(array, max_value):
    counts = np.bincount(array, minlength=max_value + 1)
    return counts[:max_value + 1]

def list_dat_files(directory):
    # 获取目录下所有以 .dat 结尾的文件
    return [directory + f for f in os.listdir(directory) if f.endswith('.dat')]
```

```python
#定义随时间变化的热流密度 q(t)
def gauss(t):
    return np.exp(-4 * np.log(2) * ((t - t0)**2) / FWHM**2)

def Ne_X(N_gamma, cross_data, rho, delta_z):
    N = N_gamma * cross_data / np.sum(cross_data) * (1 - np.exp(-np.sum(cross_data) * rho * delta_z))
    return N

def Ne_Ei(Pe, Ne_X):
    N = Pe * Ne_X
    return N

def config_to_num(config): #根据组态求索引序号
    c = 0;
    for i in range(q1):
        c += num[i + 1] * config[i]
    return c

data_config = np.array([2, 2, 6, 2, 6])
def cal_num(): #组态数累加
    num = np.zeros(20, dtype = int)
    num[q1] = 1
    for i in range(q1):
        num[q1 - i - 1] = num[q1 - i] * (data_config[q1 - i - 1] + 1)
    return num

def cal_charge(): #组态数累加
    charge = np.zeros(file_num, dtype = int)
    init_charge = np.sum(config[-1])
    charge = init_charge - np.sum(config, axis = 1)
    return charge

def list_all_config(): #枚举所有组态可能
    config = np.zeros([file_num, q1], dtype = int)
    for i in range(file_num):
        k = 0
        for j in range(q1):
            config[i][j] = (i - k) / num[j+1]
            k += num[j+1] * config[i][j]
    return config

def pl_Correspond_config_index():
    pl_Correspond_config_index = np.zeros([file_num, q1], dtype = int)
    for i in range(file_num):
        config_cache = config[i].copy()
        for j in range(q1):
            if (config_cache[j] > 0):
                config_cache[j] -= 1
                pl_Correspond_config_index[i][j] = config_to_num(config_cache)
                config_cache = config[i].copy()
    return pl_Correspond_config_index

def au_Correspond_config_index():
    au_Correspond_config_index = np.zeros([file_num, auger_size], dtype = int)
    for i in range(file_num):
        config_cache = config[i].copy()
        for j in range(au_Num[i]):
#             if (au_config[i][j][1] == au_config[i][j][2] & au_config[i][j][1] < 2):
#                 continue
#             elif (au_config[i][j][1] < 1 | au_config[i][j][1] < 1):
#                 continue
            config_cache[au_config[i][j][0]] += 1
            config_cache[au_config[i][j][1]] -= 1
            config_cache[au_config[i][j][2]] -= 1
            
            au_Correspond_config_index[i][j] = config_to_num(config_cache)
            config_cache = config[i].copy()
    return au_Correspond_config_index

def fl_Correspond_config_index():
    fl_Correspond_config_index = np.zeros([file_num, fl_size], dtype = int)
    for i in range(file_num):
        config_cache = config[i].copy()
        for j in range(fl_Num[i]):
            if (fl_config[i][j][1] < 1):
                continue
            config_cache[fl_config[i][j][0]] += 1
            config_cache[fl_config[i][j][1]] -= 1
            
            fl_Correspond_config_index[i][j] = config_to_num(config_cache)
            config_cache = config[i].copy()
    return fl_Correspond_config_index

# Numba optimized Ne_X function
#@jit(nopython=True)
def Ne_X(N_gamma, cross_data, rho, delta_z):
    N = N_gamma * cross_data / np.sum(cross_data) * (1 - np.exp(-np.sum(cross_data) * rho * delta_z))
    return N

def Ne_Ei(Pe, Ne_X):
    N = Pe * Ne_X
    return N

from concurrent.futures import ThreadPoolExecutor
# 获取 CPU 核心数
num_threads = os.cpu_count()

def process_index(j, NeX_total, pl_cross_normalize, pl_correspond_config_index, NatomX, NeX):
    NeX_single = NeX_total[j] * pl_cross_normalize[j]
    np.add.at(NatomX, pl_correspond_config_index[j], NeX_single)
    np.add.at(NeX, pl_correspond_config_index[j], NeX_single)

def parallel_update(NeX_total, pl_cross_normalize, pl_correspond_config_index, NatomX, NeX, valid_indices):
    # 设置线程池中的线程数量
    with ThreadPoolExecutor(max_workers=4) as executor:
        executor.map(
            lambda j: process_index(j, NeX_total, pl_cross_normalize, pl_correspond_config_index, NatomX, NeX),
            valid_indices
        )
```
